```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('dark')
```

```python
flights=sns.load_dataset('flights')
flights
```

```python
flights[flights['month']=='February'] 
```

```python
plt.figure(figsize=(10,8))
sns.set_style('darkgrid')
sns.lineplot(x='year',y='passengers',data=flights,hue='month')
plt.xlabel('Year')
plt.ylabel('Passengers')
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(data=tips , x='total_bill',y='tip',hue='sex' , style='smoker')
```

```python
sns.scatterplot(data=tips , x='total_bill',y='tip',hue='sex' , style='smoker',size='size')
```

```python
sns.set_style('darkgrid')
sns.scatterplot(data=tips , x='total_bill',y='tip',hue='sex',sizes=(50,300) ,size='size')
```

```python
sns.set_style('darkgrid')
sns.scatterplot(data=tips , x='total_bill',y='tip',hue='sex',sizes=(50,300) ,size='size',palette='plasma')
```

```python
sns.set_style('whitegrid')
sns.scatterplot(data=tips , x='total_bill',y='tip',hue='sex',sizes=(50,300) ,size='size',palette='Purples')
```

```python
sns.set_style('whitegrid')
sns.scatterplot(data=tips , x='total_bill',y='tip',hue='sex',style='sex',markers={'Male':'X','Female':'s'})
```

```python
sns.set_style('whitegrid')
sns.relplot(data=tips , x='total_bill',y='tip',hue='time',kind='scatter',style='sex',col='sex' ,row='day')
```

```python
sns.regplot(data=tips ,x= 'total_bill',y='tip')
```

```python
sns.regplot(data=tips ,x= 'size',y='tip',x_jitter=0.05)
```

```python
sns.regplot(data=tips ,x= 'size',y='tip')
```

```python
sns.lmplot(data=tips ,x= 'size',y='tip',x_estimator=np.mean)
```

```python
sns.lmplot(data=tips ,x= 'total_bill',y='tip',hue='smoker')
```

```python
sns.lmplot(data=tips ,x= 'total_bill',y='tip',hue='smoker',col='smoker')
```

```python
sns.lmplot(data=tips ,x= 'total_bill',y='tip',hue='smoker',col='smoker',row='sex')
```

```python
sns.lmplot(data=tips ,x= 'total_bill',y='tip',hue='smoker',col='smoker',row='sex',truncate=False)
```

```python
sns.lmplot(data=tips ,x= 'total_bill',y='tip',hue='smoker',col='smoker',row='sex',scatter_kws={'s': 100,'edgecolor':'blue'})
```
